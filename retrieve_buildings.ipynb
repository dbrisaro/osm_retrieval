```python
# import libraries
import osmnx as ox
import numpy as np
import matplotlib.pyplot as plt
```

```python

cities = [
    "Kingston, Jamaica",
    "Port of Spain, Trinidad and Tobago",
    "Bridgetown, Barbados",
    "Santo Domingo, Dominican Republic",
    "Castries, Saint Lucia"
]
```

```python
def get_buildings(place, save=None):
    """
    Fetch building footprints from OpenStreetMap for a given place.

    Parameters
    ----------
    place : str
        Place name recognized by OSM (e.g., "Castries, Saint Lucia").
    save : str, optional
        Directory to save output. If None, nothing is written to disk.

    Returns
    -------
    GeoDataFrame
        Buildings with columns: building, name, addr:city, geometry.
    """
    import osmnx as ox

    tags = {"building": True}
    gdf = ox.features.features_from_place(place, tags)
    gdf = gdf[["building", "name", "addr:city", "geometry"]]

    if save:
        fname = place.replace(", ", "_").replace(" ", "_").lower()
        gdf.to_file(f"{save}/buildings_{fname}.gpkg", driver="GPKG")

    return gdf
```

```python
place = cities[-1]
df = get_buildings(place)
```
